```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
## To check data types
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
```

```python
# help(plt.pie)
```

```python
# Pie Chart Top3 countries that uses Zomato
plt.pie(x=country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation : Maximum no of transactions of Zomato is made in India and then US and then UK

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings.head(ratings.size)
```

Here by seeing we can say the max no of people i.e 2148 are not giviing rating

***Observations:***
- When rating is between 4.5 to 4.9 --->Excelleny
- When rating is between 4.0 to 4.4 --->Very good
- When rating is between 3.5 to 3.9 --->good
- When rating is between 3.0 to 3.4 --->Average
- When rating is between 2.5 to 2.9 --->Average
- When rating is between 2.0 to 2.4 --->Poor


```python
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
_=sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings,palette=['black','red','orange','yellow','green','green'])
```

**Observations**:
  1. Not Rating count is very high
  2. Maximum no of ratinngs are between 2.5 to 3.4
  
If we have missing values in the ratings then we can impute the average between this range because maximum ratings are occured in this range

```python
## Count plot
sns.countplot(x='Rating color',data=ratings,palette=['black','red','orange','yellow','green','green'])
```

These Y axis is not wrto count it is no of times the color is repeated

```python
### Find the countries names that have given 0 rating
final_df.groupby(['Country','Rating color']).size()
```

```python
final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index().rename(columns={0:'White Rating Count'}).sort_values(by='White Rating Count',ascending=False)
```
