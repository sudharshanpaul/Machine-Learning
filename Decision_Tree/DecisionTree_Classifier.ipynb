#### Decision Tree Classifier Practical Implementation

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris
```

```python
iris['DESCR']
```

```python
print(iris['target_names'])
print(iris['target'])
```

```python
# Independent Features
X = pd.DataFrame(iris['data'],columns=iris['feature_names'])
X.head()
```

```python
# Target feature or Dependent
y = iris['target']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

**In Decision Tree we don't require Standardization or Normalization**
- We'll directly apply the algorithm

```python
from sklearn.tree import DecisionTreeClassifier
treeClassifier = DecisionTreeClassifier()
```

```python
treeClassifier.fit(X_train,y_train)
```

```python
y_pred = treeClassifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
## Visualize the Decision Tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeClassifier,filled=True)
```

- We can see that the max Classification is done at the depth 2. So to avoid overfitting let's use **Post Pruning**

```python
treeClassifier = DecisionTreeClassifier(max_depth=2)
treeClassifier.fit(X_train,y_train)
## Visualize the Decision Tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeClassifier,filled=True)
```

```python
y_pred = treeClassifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
