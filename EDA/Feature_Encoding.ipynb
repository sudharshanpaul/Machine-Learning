# Feature Encoding

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
df = pd.read_csv('/content/Churn_Modelling.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.drop(columns=['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

Let's do Encoding for the Gender column

#####Handling Missing Values

```python
df.Gender.value_counts()
```

```python
df.Gender.mode()[0]
```

```python
df.Gender = df.Gender.fillna(df.Gender.mode()[0])
```

```python
df.Gender.value_counts()
```

```python
df.info()
```

## Label Encoding

```python
le = preprocessing.LabelEncoder()
df['Gender_label'] = le.fit_transform(df.Gender)
df.head(5)
```

```python
df.Gender_label.value_counts()
```

```python
df['Geography_label'] = le.fit_transform(df.Geography)
df.head(5)
```

```python
df.Geography_label.value_counts()
```

```python
df.drop(columns=['Geography_label'],axis=1,inplace=True)
```

```python
df.columns
```

## One Hot Encoding

```python
one_hot.head(5)
```

```python
df.columns
```

```python
df_dummies = pd.get_dummies(df,dtype=int)
df_dummies.head(5)
```

## Target Encoding

```python
pip install category_encoders
```

```python
from category_encoders import TargetEncoder
```

```python
t_encoder = TargetEncoder()
df1 = pd.read_csv('/content/Churn_Modelling.csv')
```

```python
df1.drop(columns=['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
df1['Gender'] = df1.Gender.fillna(df1.Gender.mode()[0])
```

```python
df1['Gender Encoded'] = t_encoder.fit_transform(df1.Gender,df1.Exited)
df1.head(5)
```

Now we can use this Gender Encoded feature instead of Gender and this is only done when there is target variable

```python
df1['Gender Encoded'].value_counts()
```

***Benifits of Target Encoding***
- Target encoding is a simple and quick encoding method that doesn't add to the dimensionality of the dataset
Therefore, it may be used as a good first try technique

***Limitations***
- It is dependent on the target distributionn which means target encoding requires careful validation as it can be prone to overfitting.
