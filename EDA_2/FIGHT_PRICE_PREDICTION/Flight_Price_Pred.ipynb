```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import openpyxl
```

```python
train_df = pd.read_excel('Data_Train.xlsx')
train_df.head()
```

```python
test_df = pd.read_excel('Test_set.xlsx')
test_df.head()
```

```python
df = pd.concat([train_df,test_df],ignore_index=True)
df.head()
```

```python
## Handling Date of journey here my aim is to seperate date,month and year
df['Date_of_Journey'].str.split('/').str[0]
```

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head(2)
```

```python
df.info()
```

```python
## Date, Month, Year are the features which are in categorical form. Now we have to connvert into numerical form
df['Date']=df['Date'].astype(int)
df['Month'] = df["Month"].astype(int)
df['Year'] = df["Year"].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df["Arrival_Time"].apply(lambda x:x.split(" ")[0]).str.split(":").str[1]
```

```python
df['Arrival_hour'] = df['Arrival_Time'].str.split(":").str[0]
df['Arrival_min'] = df["Arrival_Time"].apply(lambda x:x.split(" ")[0]).str.split(":").str[1]
```

```python
df.head(2)
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
```

```python
df.info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df['Dep_hour']=df['Dep_Time'].str.split(":").str[0]
df['Dep_min']=df['Dep_Time'].str.split(":").str[1]
```

```python
df.head(2)
```

```python
df['Dep_hour']=df['Dep_hour'].astype(int)
df['Dep_min']=df['Dep_min'].astype(int)
```
