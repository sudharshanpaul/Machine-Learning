# Feature Scaling

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Churn_Modelling.csv')
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

***What is Normalization?***

Normalization is a scaling technique in which the values are shifted and rescaled so that they end up ranging between 0 and 1. It is also known as Min-Max Scaling

***What is Standardization?***

Standardization is another scaling technique where the values are centered around the mean with a unit standard deviation. This means that the mean of the attribute becomes zero and the resultant distribution has unit Standard deviation

```python
df.head()
```

```python
df.describe().round(2) # which shows the statistical information of all the numerical values
```

*****NORMALIZATION*****

```python
df.head(5)
```

```python
# Creating a new dataframe with the Age and Tenuew where we want to do the feature scaling using Normalization
new_df = pd.DataFrame(df,columns = ['Age','Tenure'])
```

```python
new_df.head(5)
```

```python
new_df.Age = new_df.Age.fillna(new_df.Age.mean())
new_df.Tenure = new_df.Tenure.fillna(df.Tenure.mean())
```

```python
new_df.isnull().sum()
```

```python
scaler = MinMaxScaler() # Instantiating the MinMaxScaler() function
normalized_df = scaler.fit_transform(new_df)
print(normalized_df)
```

```python
x_array = np.array([[2],[3],[5],[6],[6]])
scaler = MinMaxScaler()
normalized_arr = scaler.fit_transform(x_array)
normalized_arr
```

*****STANDARDIZATION*****

```python
scaler = StandardScaler()
standardized_df = scaler.fit_transform(new_df)
print(standardized_df)
```

```python
x_array = np.array([[2],[3],[5],[6],[6]])
scaler = StandardScaler()
standardized_arr = scaler.fit_transform(x_array)
standardized_arr
```
