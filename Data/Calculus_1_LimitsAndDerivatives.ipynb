In linear algrbra we found the line equation using Moore Pendrose PseudoInverse
But here we are approaching this problem with Calculus and regression

Now we are let fitting the slope of the line as m = 0.9

```python
m = torch.tensor([0.9]).requires_grad_()
m
```

Let's set the y-intercept as 0.1 as commonly ad test it

```python
b = torch.tensor([0.1]).requires_grad_()
b
```

```python
def regression_plot(x,y,m,b):
  fig, ax = plt.subplots()
  ax.scatter(x,y)

  x_min,x_max = ax.get_xlim()
  y_min,y_max = m.item()*x_min + b.item(),m.item()*x_max + b.item()
  ax.set_xlim(x_min,x_max)
  ax.plot([x_min,x_max],[y_min,y_max],c='red')
```

```python
regression_plot(x,y,m,b)
```

```python
def regression(x,m,b):
  return m * x + b
```

## Machine Learning

To fit that line correctly lets use a ""Machine Learning"
- 4 Easy step process :)

**Step 1**: Forward Pass

```python
yhat = regression(x,m,b)
yhat
```

**Step 2**: Compares yhat with the original y values so that we can calculate the cost *C*
There is a Pytorch MSELoss method but let's define how it works. MSE cost is defined by
https://chatgpt.com/c/6e3b8264-f475-41f0-aa95-773316cc7ff7

```python
def mse(my_yhat,my_y):
  sigma = torch.sum((my_yhat-my_y)**2)
  return sigma/len(my_y)
```

```python
C = mse(yhat,y)
C
```

**Step 3**: Use Autodiff to calcuate the gradient of C w.r.to parameters

```python
C.backward()
```

```python
m.grad    # As the slope of C w.r.to m is positive so that by reducing the value of m we can reduce the Cost
# As the the gradient with m is soo high which meand the decrease in m will show greater impact in C
```

```python
b.grad
```

**Step 4** : Gradient Descent

```python
# Here the optimizer works based on the Cost gradient with respect to the parameters and adjusts those parameters accordingly to fit the equation
optimizer = torch.optim.SGD([m,b],lr=0.01)
```

```python
optimizer.step()
```

```python
m
```

```python
b
```

```python
regression_plot(x,y,m,b)   # For one single step of modifying the values of m and b the line is changed in a big manner
```

Now repeat the step 1 and step 2 to confirn the cost is decreases or not

```python
C = mse(regression(x,m,b),y)
C   # We can see the Cost is decreased
```

Put the 4 steps in a loop to iteratively to minimize the cost to zero

```python
epochs = 1000
for epoch in range(epochs):
  optimizer.zero_grad()  # Resets the gradient to zero else it accumulates the space

  yhat = regression(x,m,b)   #Step1 : Calculating the yhat
  C = mse(yhat,y)            #Step2 : Finding the cost or the loss
  C.backward()               #Step3 : Finding how the cost is varying with m and b
  optimizer.step()           #Step4 : Gradient Descent

  print('Epoch {},cost {},m grad {},b grad {},'.format(epoch,'%.3g' % C.item(), '%.3g' % m.grad.item(),'%.3g' % b.grad.item()))
```

```python
#Step 1
m = torch.tensor([0.3]).requires_grad_()
m
```

```python
b = torch.tensor([10.0]).requires_grad_()
b
```

```python
def regression_plt(x,y,m,b):

  fig, ax = plt.subplots()
  ax.scatter(x,y)

  x_min,x_max = ax.get_xlim()
  y_min, y_max = m.item()*x_min + b.item(),m.item()*x_max + b.item()
  ax.set_xlim(x_min,x_max)
  ax.plot([x_min,x_max],[y_min,y_max],c='red')

```

```python
def regression_val(x,m,b):
  return m*x + b
```

```python
regression_plt(x,y,m,b)
```

```python
def cal_mse(yhat,y):
  sum = torch.sum((yhat-y)**2)
  return sum/len(y)
```

```python
# Step 3 Calculating the cost
yhat = regression_val(x,m,b)
```

```python
C = cal_mse(yhat,y)
C
```

```python
C.backward()
```

```python
m.grad
```

```python
b.grad
```

```python
# Step 4 Gradient descent
optimizer = torch.optim.SGD([m,b],lr=0.01)
```

```python
optimizer.step()
```

```python
b
```

```python
C = cal_mse(regression_val(x,m,b),y)
C
```

```python
# Repeat this steps for more no of times so that the model gives good values of m and b

epochs = 1000

for epoch in range(epochs):
  optimizer.zero_grad()

  yhat = regression_val(x,m,b)
  C = cal_mse(yhat,y)
  C.backward()
  optimizer.step()

  print('Epoch {},cost {},m grad {},b grad {},'.format(epoch,'%.3g' % C.item(), '%.3g' % m.grad.item(),'%.3g' % b.grad.item()))
```
